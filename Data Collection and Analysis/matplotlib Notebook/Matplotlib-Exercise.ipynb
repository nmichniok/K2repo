```python
#5 Custom tick marks and labels
plt.plot(x, y1, 'c-.', lw = 4.0)
plt.plot(x, y2, 'm:')
plt.axis([-6,6,-1,1])
plt.xticks([-2*np.pi, (-3/2)*np.pi, -np.pi, -np.pi/2, np.pi, (3/2)*np.pi, 2*np.pi],
          [r'$-2\pi$', r'$-(3/2)\pi$', r'$-\pi$', r'$\pi/2$', r'$\pi$', r'$(3/2)\pi$', r'$2\pi$'])
plt.yticks([-1, -0.5, 0.0, 0.5, 1.0], 
          [r'$-1$', r'$-0.5$', r'$0$', r'$0.5$', r'$1$'])
plt.grid(True)
plt.show()
```

```python
#6 plot spines centered at origin
plt.plot(x, y1, 'c-.', lw = 4.0)
plt.plot(x, y2, 'm:')
plt.axis([-6,6,-1,1])
plt.xticks([-2*np.pi, (-3/2)*np.pi, -np.pi, -np.pi/2, np.pi, (3/2)*np.pi, 2*np.pi],
          [r'$-2\pi$', r'$-(3/2)\pi$', r'$-\pi$', r'$\pi/2$', r'$\pi$', r'$(3/2)\pi$', r'$2\pi$'])
plt.yticks([-1, -0.5, 0.0, 0.5, 1.0], 
          [r'$-1$', r'$-0.5$', r'$0$', r'$0.5$', r'$1$'])
#plt.grid(True)

ax = plt.gca()  #gca = get current axis in order to edit
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.show()
```

```python
#7 Add legend and labels
plt.plot(x, y1, 'c-.', lw = 4.0, label = r'$\sin(x)$')
plt.plot(x, y2, 'm:', lw=2.0, label = r'$\cos(x)$')
plt.axis([-6,6,-1,1])
plt.xticks([-2*np.pi, (-3/2)*np.pi, -np.pi, -np.pi/2, np.pi, (3/2)*np.pi, 2*np.pi],
          [r'$-2\pi$', r'$-(3/2)\pi$', r'$-\pi$', r'$\pi/2$', r'$\pi$', r'$(3/2)\pi$', r'$2\pi$'])
plt.yticks([-1, -0.5, 0.0, 0.5, 1.0], 
          [r'$-1$', r'$-0.5$', r'$0$', r'$0.5$', r'$1$'])
#plt.grid(True)
plt.legend(loc='upper right', frameon=False)
ax = plt.gca()  #gca = get current axis in order to edit
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.show()
```

```python
#8 Add 2 more data sets, adjust legend and labels
y3 = np.sin(x) + np.sin(2*x)
y4 = np.cos(x) + np.cos(2*x)


plt.plot(x, y1, 'c-.', lw = 4.0, label = r'$\sin(x)$')
plt.plot(x, y2, 'm:', lw=2.0, label = r'$\cos(x)$')

plt.plot(x, y3, color='blue', lw=1.0, ls='--',label = r'$\sin(x) + \sin(2x)$')
plt.plot(x, y4, color='red', lw=3.0, label = r'$\cos(x) + \cos(2x)$')

plt.axis([-6,6,-4,4])
plt.xticks([-2*np.pi, (-3/2)*np.pi, -np.pi, -np.pi/2, np.pi, (3/2)*np.pi, 2*np.pi],
          [r'$-2\pi$', r'$-(3/2)\pi$', r'$-\pi$', r'$\pi/2$', r'$\pi$', r'$(3/2)\pi$', r'$2\pi$'])
plt.yticks([-4, -2, 0.0, 2, 4.0], 
          [r'$-4$', r'$-2$', r'$0$', r'$2$', r'$4$'])
#plt.grid(True)
plt.legend(loc='lower right', frameon=False)
ax = plt.gca()  #gca = get current axis in order to edit
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.title('Lines')

plt.show()
```

```python
import pandas as pd
```

```python
#1
crash_df = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/gamclass/airAccs.csv", index_col=0)
crash_df.head()
```

```python
#2
plt.hist(crash_df.Dead.dropna(), bins=25)
plt.axis([0,300,0,5000])
plt.xticks(np.arange(0,300,25))
plt.title('Deaths in Air Crashes')
plt.ylabel('Count')
plt.xlabel('Deaths')
plt.show()
```

```python
#3
ts = crash_df.copy()  #copy data frame 
ts = ts.set_index(pd.DatetimeIndex(ts.Date)).drop('Date', axis=1) #set Date field as the index

t_range = pd.date_range(start=crash_df.Date[1], end=crash_df.Date[len(crash_df)]) #build list of range of dates
ts = ts.loc[t_range]
ts = ts.fillna(0)
ts['year'] = ts.index.year
#ts['month'] = ts.index.month
ts_yearly_tot = ts.groupby('year').sum().Dead #create subset query summing Dead count by year
#ts_yearly_tot.head(10)
```

```python
plt.plot(ts_yearly_tot)
plt.xlim(ts_yearly_tot.index[0], ts_yearly_tot.index[len(ts_yearly_tot.index) - 1])
plt.title('Airline Deaths by Year')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.show()
```

```python
#4
pass_df_raw = pd.read_csv("AirPass.csv", encoding='ISO-8859-1')
pass_df_raw.head()
```

```python
pass_df = pass_df_raw.iloc[:,4:] #keep all the rows, but get rid of the first 4 columns
pass_df = pass_df.sum() #sum all the rows 
#pass_df.head(20)
pass_df = pd.DataFrame({'year': pass_df.index.astype(float), 'passengers': pass_df})
pass_df.head()

data_acc_temp = crash_df[['Date', 'Dead']] #create temp table from crash_df of just date field and dead count field
data_acc_temp = data_acc_temp.set_index(pd.DatetimeIndex(crash_df.Date)) #set index of new temp table to date as pd.datetimeindex format
data_acc_temp['year'] = data_acc_temp.index.year #create new column for year from the date index field just made
data_acc_temp = data_acc_temp.groupby('year').sum() #write over temp table with new table that sums the Dead field over the year field
data_acc_temp.head()
data_acc_temp = pd.DataFrame({'year': data_acc_temp.index.astype(float), 'dead': data_acc_temp.Dead})

#data_all = pd.merge(left=data_acc_temp, right=pass_df, how='inner', on='year')
data_all = pd.merge(pass_df, data_acc_temp)[['year', 'passengers', 'dead']]
data_all = data_all[(data_all.passengers != 0) & (data_all.dead != 0)]
data_all.head()
```

```python
#5
data_all['deaths_perM_pass'] = (data_all.dead / data_all.passengers) * 1000000 #create new field in data_all that shows dead per passengers in millions
data_all['year'] = [int(y) for y in data_all.year]

```

```python
#5 Two graphs visually representing Deaths and passengers over time

plt.figure(1)
plt.subplot(311)
plt.plot(data_all.year, data_all.passengers / 1000000)
#plt.ylim(0,3e10)
plt.ylabel('Total Passengers \n(,000,000s)')
plt.xlabel('Year')

plt.figure(2)
plt.subplot(312)
plt.plot(data_all.year, data_all.dead)
plt.ylabel('Total Deaths')
plt.xlabel('Year')

plt.figure(3)
plt.subplot(313)
plt.plot(data_all.year, data_all.deaths_perM_pass)
plt.ylabel('Deaths per \nMillion')
plt.xlabel('Year')

plt.show()
```

```python
#6 Make a pie chart representing the number of deaths for each decade.
data_all_d = data_all #make a copy of the data frame
data_all_d['decade'] = data_all_d.year // 10 * 10 #create a new field, decade
data_all_d = data_all_d.groupby('decade').sum()[['dead']] #create new data field that has the number of dead summed by decade

labels = '1970', '1980', '1990', '2000', '2010'
colors = ['gold', 'hotpink', 'beige', 'lightskyblue', 'green']
plt.pie(data_all_d.dead, colors=colors, labels=labels, autopct='%1.1f%%', shadow=True, startangle=50)
plt.title('Airline Deaths Over the Decades')
plt.axis('equal')
plt.legend(labels, loc='best')
plt.show()
```
